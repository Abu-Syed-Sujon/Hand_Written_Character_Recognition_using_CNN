```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('E:/Dataset/A_Z Handwritten Data.csv').astype('float32')
data.head(10)
```

```python
X = data.drop('0',axis = 1)
y = data['0']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

x_train = np.reshape(x_train.values, (x_train.shape[0], 28,28))
x_test = np.reshape(x_test.values, (x_test.shape[0], 28,28))

print("Shape of Training data: ", x_train.shape)
print("Shape of Testing data: ", x_test.shape)
```

```python
import cv2
shuffle_data = shuffle(x_train)

fig, axes = plt.subplots(3,3, figsize = (10,10))
axes = axes.flatten()

for i in range(9):
    _, shu = cv2.threshold(shuffle_data[i], 30, 200, cv2.THRESH_BINARY)
    axes[i].imshow(np.reshape(shuffle_data[i], (28,28)), cmap="Greys")
plt.show()
```

```python
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1],x_train.shape[2],1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2],1)

print("New shape of training data: ", x_train.shape)
print("New shape of testing data: ", x_test.shape)
```

```python
import tensorflow
from tensorflow.keras.utils import to_categorical

y_training = to_categorical(y_train, num_classes = 26, dtype='int')
y_testing = to_categorical(y_test, num_classes = 26, dtype='int')

print("New shape of training labels: ", y_training.shape)
print("New shape of testing labels: ", y_testing.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python
model = Sequential()

model.add(Conv2D(64 , (3, 3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPool2D(2, 2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(2, 2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Flatten())

model.add(Dense(128,activation ="relu"))
model.add(Dense(256,activation ="relu"))

model.add(Dense(26,activation ="softmax"))
```

```python
model.compile(optimizer = Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

# 16. To visualize our model on Customize image:

```python
import os
import cv2
import tensorflow as tf
import numpy as np
from keras.models import load_model

# Load trained model
model = load_model(r'E:\Jupyter Notebook\Machine Learning\handwritten-character-recognition-source-code\HandWritten_best_epochweights.10-0.10.h5')

# Character labels
words = {i: chr(65 + i) for i in range(26)}  # Maps indices to A-Z

# Load and preprocess image
image = cv2.imread('G.png')
if image is None:
    raise FileNotFoundError("Image 'G.png' not found. Check the path.")

image_copy = image.copy()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (400, 440))

# Convert to grayscale and apply threshold
image_copy = cv2.GaussianBlur(image_copy, (7, 7), 0)
gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
_, img_thresh = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY_INV)

# Resize and normalize for model input
final_image = cv2.resize(img_thresh, (28, 28))
final_image = np.reshape(final_image, (1, 28, 28, 1))
final_image = final_image.astype('float32') / 255.0  # Normalize

# Model prediction
prediction_index = np.argmax(model.predict(final_image), axis=1)[0]
prediction = words.get(prediction_index, "?")  # Get predicted character

# Display result
cv2.putText(image, f"Prediction: {prediction}", (20, 410), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
cv2.imshow('Handwritten Character Recognition', image)

# Wait for ESC key to close
while True:
    k = cv2.waitKey(1) & 0xFF
    if k == 27:  # ESC key
        break
cv2.destroyAllWindows()
```
